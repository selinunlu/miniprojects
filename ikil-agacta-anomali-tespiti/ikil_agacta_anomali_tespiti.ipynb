# 3. İkil Ağaçta Anomali Tespiti

* İkil ağaçların (binary tree) kullanıldığı bir yazılım sistemi üzerinde çalıştığınızı varsayalım. Yanlış programlama nedeniyle aynı elemanı (node) gösteren birden fazla elemana sahip ağaçlar sisteme kaydedilmiş. Elemanlar üzerinde bağlar (link) dışında başka bir bilgi tutulmuyor. 


* Bir ikil ağaçta böyle bir anomali varsa `“True”`, yoksa `“False“` değerini döndüren işlevi yazınız.

```python
parent1 = ["A","A","C","C","D"];            
child1  = ["B","C","D","E","F"] 
```

```python
parent2 = ["A","A","B","C","C","D","E"];    
child2  = ["B","C","D","D","E","F","F" ]
```

```python
class BinaryTree:    
              
    def __init__(self, child, parent):
        self.Child = child
        self.Parent = parent
        self.d = {}            
              
    def is_anomaly(self):         
           
        for i in range(len(self.Child)):
            self.d[self.Child[i]] = self.Parent[i]

        if len(self.d) == len(self.Child):
            return False
        else:
            return True
```

```python
tree1 = BinaryTree(child1,parent1)
tree1.is_anomaly()
```

```python
tree2 = BinaryTree(child2,parent2)
tree2.is_anomaly()
```
