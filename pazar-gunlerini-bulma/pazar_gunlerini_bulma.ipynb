# 2. Pazar Günlerini Bulma

* `1.1.1900` ile `31.12.2000` tarihleri arasında, `ayın ilk günü` `pazar` olan tarihleri listeleyen uygulamayı yazınız.


* **İpucu:** 1.1.1900 tarihinin “Pazartesi” olduğu bilgisini kodunuzun performansını arttırmak için kullanınız.

```python
def hangi_gun(date):
        
    gun, ay, yil = date.split(".")

    if int(ay) == 1:
        b = 1
    elif int(ay) == 2:
        b = 4
    elif int(ay) == 3:
        b = 4
    elif int(ay) == 4:
        b = 7
    elif int(ay) == 5:
        b = 2   
    elif int(ay) == 6:
        b = 5   
    elif int(ay) == 7:
        b = 7
    elif int(ay) == 8:
        b = 3   
    elif int(ay) == 9:
        b = 6   
    elif int(ay) == 10:
        b = 1
    elif int(ay) == 11:
        b = 4   
    elif int(ay) == 12:
        b = 6

    return ( ((int(gun) % 7) + int(b) + (((int(yil)-1900) % 4) % 7)) %7 )  
```

```python
istenen_gunler=[]

for yil in range(1900,2001,1):    
    for ay in range(1,13,1):        
        gun = 1          
        date = ".".join([str(gun), str(ay), str(yil)])

        if (hangi_gun(date) == 1):
            istenen_gunler.append(date)                        
```
